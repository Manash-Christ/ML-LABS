```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/insurance.csv')
df.info()
```

```python
dfd = pd.get_dummies(df, drop_first=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
num = [x for x in df if str(df[x].dtype) != 'object']
```

```python
mx = MinMaxScaler()
for i in num:
  dfd[i] = mx.fit_transform(dfd[i].array.reshape(-1,1))
```

```python
dfd
```

```python
df[num].hist()
```

```python
num.remove('charges')
```

```python
for i in num:
  sns.scatterplot(data=dfd, x=dfd[i], y = dfd['charges'], hue = df.smoker, size=df.region)
  plt.show()
```
